```python
#lookup table mapping 
import pandas as pd

lookup_df = pd.read_csv("D:\VIS coursework\Lookup_table.csv")

lookup_df = lookup_df.rename(columns={
    'LAD11CD': 'Old_Code',
    'LAD21CD': 'New_Code'
})

def map_area_codes(df):
    if 'Area Code' in df.columns:
        df = df.merge(lookup_df[['Old_Code', 'New_Code']], how='left', left_on='Area Code', right_on='Old_Code')
        df['Area Code'] = df['New_Code'].combine_first(df['Area Code'])  # use 2021 code if available
        df.drop(columns=['Old_Code', 'New_Code'], inplace=True)
    return df
```

```python

df = pd.read_excel("D:\\VIS coursework\\excel datasets 2011 VISCOU\\highest level of qualification_2011.xlsx",sheet_name = "Data",skiprows=7).iloc[:-60]
```

```python
df.replace('-', pd.NA, inplace=True)
df.rename(columns={
    df.columns[0]:"Region",
    df.columns[1]:"Area Code"
},inplace=True)
```

```python
df.isnull().sum()
```

```python
target_cols = ['Apprenticeship', 'Other qualifications']
non_feature_cols = ['local authority: district / unitary (prior to April 2015)']
features = df.drop(columns=target_cols + non_feature_cols, errors='ignore')
features = features.apply(pd.to_numeric, errors='coerce')
```

```python
features
```

```python

#baysien methods 
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module='openpyxl')


import pandas as pd
from sklearn.linear_model import BayesianRidge
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features)

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)
for col in target_cols:
    target = pd.to_numeric(df[col], errors='coerce')
    not_missing = target.notna()

    model = BayesianRidge()
    model.fit(features_scaled[not_missing], target[not_missing])

    missing_mask = ~not_missing
    predicted = model.predict(features_scaled[missing_mask])
    
    df.loc[missing_mask, col] = predicted.clip(min=0)


df.to_excel("highest_level_of_qualification_2011_imputed_cleaned.xlsx", index=False)
print(" Bayesian done")
df
```

```python
def pivot_df(df,dim_col):
    id_vars = ['Region', 'Area Code', 'Year']
    value_vars = [col for col in df.columns if col not in id_vars]
        long_df = df.melt(id_vars=id_vars, value_vars=value_vars,
                      var_name=dim_col, value_name="Counts")
    
    return long_df


merged_df = pivot_df(merged_df,"Industry")
df_combined= pivot_df(df_combined,"Economic Activity")
df_qualification= pivot_df(df_qualification,"Qualifications")



```

```python
#pca
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df = pd.read_excel("D:\VIS coursework\cleaned_data\merged_cleaned_industry_employment.xlsx")

numeric_df = df.select_dtypes(include='number')
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_data)

df['PCA1'] = pca_components[:, 0]
df['PCA2'] = pca_components[:, 1]

df.to_excel("industry_employment_with_pca.xlsx", index=False)

print(" PCA done")
```

```python
#t-SNE
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

df = pd.read_excel("D:\VIS coursework\cleaned_data\highest_level_of_qualification_combined_cleaned.xlsx")  # Change filename here
numeric_df = df.select_dtypes(include='number')

scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=42)
tsne_components = tsne.fit_transform(scaled_data)
df['TSNE1'] = tsne_components[:, 0]
df['TSNE2'] = tsne_components[:, 1]
df.to_excel("qualification_with_tsne.xlsx", index=False)

print(" t-SNE done")
```
